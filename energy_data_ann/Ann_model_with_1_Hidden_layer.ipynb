```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv(r"E:\Data Science by SRK\Deep Learning\energy_data\cleaned_dataset_energy.csv")
```

```python
df
```

# X and y

```python
X = df.drop(columns = ['Energy'])
y = df['Energy']
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = True)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# ANN_MODEL

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def ann_model_regressor():
    model = Sequential()

    # adding first hidden layer
    model.add(Dense(input_dim = 4, units = 8, activation = 'relu', kernel_initializer = 'uniform'))

    # adding output layer
    model.add(Dense(units = 1, activation = 'relu', kernel_initializer = 'uniform'))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
```

```python
ann = ann_model_regressor()
```

```python
## Trainin the ANN model on the Training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
## Predicting the Result
ypred_train = ann.predict(X_train)
ypred_test = ann.predict(X_test)
```

```python
# forword
from sklearn.metrics import mean_squared_error
print("MSE for train data : ", mean_squared_error(y_train, ypred_train))
print("MSE for test data : ", mean_squared_error(y_test, ypred_test))

```

```python
# backword
from sklearn.metrics import r2_score
print("R2 for train data : ", r2_score(y_train, ypred_train))
print("R2 for test data : ", r2_score(y_test, ypred_test))

```

# Cross validation

```python
def build_cross_regressor():
    regressor = Sequential()
    regressor.add(Dense(input_dim=4, units=8, activation='relu', kernel_initializer='uniform'))
    regressor.add(Dense(units=1, activation='relu', kernel_initializer='uniform'))

    regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return regressor 
```

```python
from scikeras.wrappers import KerasRegressor
regressor = KerasRegressor(build_fn=build_cross_regressor, batch_size=32, epochs=100)
```

```python
from sklearn.model_selection import cross_val_score
accuracies_1 = cross_val_score(regressor, X , y, cv= 5)
```

```python
print(accuracies_1)
accuracies_1.mean()
```
