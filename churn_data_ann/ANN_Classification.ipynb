# X and y

```python
X = df.drop(columns = ['Exited'])
y = df['Exited']
```

```python
X
```

```python
y
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = True)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Modelling

# perceptron model

**Without Hidden layer**

```python
from keras.models import Sequential
Perc_model = Sequential()

from keras.layers import Dense
Perc_model.add(Dense(input_dim = 11, units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# compiling ANN : what it will do in forword propogation

Perc_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
Perc_model.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
# Prediction & Evaluation
```

```python
y_pred_train = Perc_model.predict(X_train)

y_pred_test = Perc_model.predict(X_test)
ypred_train = (y_pred_train > 0.5)
ypred_test = (y_pred_test > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print("Train accuracy : ", accuracy_score(y_train, ypred_train))

print("Test accuracy : ", accuracy_score(y_test, ypred_test))
print("confusion_matrix : ", confusion_matrix(y_test, ypred_test))
```

**This model is giving good accuray**

# cross validation perceptron

```python
def build_cross_classifier():
    classifier = Sequential()
    classifier.add(Dense(input_dim=11, units=1, activation='sigmoid', kernel_initializer='uniform'))

    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics =['accuracy'])

    return classifier 
```

```python
from scikeras.wrappers import KerasClassifier

classifier = KerasClassifier(build_fn=build_cross_classifier, batch_size=32, epochs=100)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier, X , y, cv= 5)
```

```python
print(accuracies)
accuracies.mean()
```
